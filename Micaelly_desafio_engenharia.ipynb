```python
import pandas as pd
```

```python
df = pd.read_fwf('pacotes.txt', widths=[3] * 5, names=['regiao_de_origem', 'regiao_de_destino', 'codigo_da_loggi', 'codigo_do_vendedor', 'tipo_do_produto'])
df.index = pd.Index(
    df.apply(lambda x: f"{x['regiao_de_origem']:03}{x['regiao_de_destino']:03}{x['codigo_da_loggi']:03}{x['codigo_do_vendedor']:03}{x['tipo_do_produto']:03}", axis=1), 
    name='pacote')
df.head()
```

```python
df['valido'] = True
df
```

Invalidando entregas de produto com tipo diferente dos fornecidos

```python
codigos_de_produtos_validos = {1, 111, 333, 555, 888}
codigos_de_produtos_validos
```

```python
df['valido'] &= df['tipo_do_produto'].isin(codigos_de_produtos_validos)
df
```

Verificando se é jóia de origem no centro-oeste

```python
codigo_joia = 1
codigos_centro_oeste = range(201, 300)
```

```python
~((df['tipo_do_produto'] == codigo_joia) & (df['regiao_de_origem'].isin(codigos_centro_oeste)))
```

```python
df['valido'] &= ~((df['tipo_do_produto'] == codigo_joia) & (df['regiao_de_origem'].isin(codigos_centro_oeste)))
df
```

O vendedor 367 está inativo, invalidando os pacotes dele

```python
vendedor_inativo = 367
```

```python
df['valido'] &= df['codigo_do_vendedor'] != vendedor_inativo
df
```

```python
def verifica_regiao(codigo_regiao):
    if codigo_regiao in range(201,300):
        return "Centro-oeste"
    elif codigo_regiao in range(300,400):
        return "Nordeste"
    elif codigo_regiao in range (400,500):
        return "Norte"
    elif codigo_regiao in range(1,100):
        return "Sudeste"
    elif codigo_regiao in range(100,200):
        return "Sul"
    else:
        return None
    
assert verifica_regiao(5) == 'Sudeste'
assert verifica_regiao(105) == 'Sul'
assert verifica_regiao(205) == 'Centro-oeste'
assert verifica_regiao(305) == 'Nordeste'
assert verifica_regiao(405) == 'Norte'
assert verifica_regiao(0) is None
print('Função ok')
```

Removendo pacotes com códigos de origem e destino inválidos

```python
df['nome_regiao_de_destino'] = df['regiao_de_destino'].apply(verifica_regiao)
df['nome_regiao_de_origem'] = df['regiao_de_origem'].apply(verifica_regiao)

df['valido'] &= df['nome_regiao_de_origem'].notnull() & df['nome_regiao_de_destino'].notnull()
df
```

# Questão 6
> Gerar o relatório/lista de pacotes por destino e por tipo (Considere
apenas pacotes válidos);

```python
def verifica_tipo_do_produto(codigo_produto):
    if codigo_produto == 1 :
        return "joias"
    elif codigo_produto == 111:
        return "livros"
    elif codigo_produto == 333:
        return "eletronicos"
    elif codigo_produto == 555:
        return "bebidas"
    elif codigo_produto == 888:
        return "brinquedos"
    else:
        return None
    
assert verifica_tipo_do_produto(1) == 'joias'
assert verifica_tipo_do_produto(111) == 'livros'
assert verifica_tipo_do_produto(333) == 'eletronicos'
assert verifica_tipo_do_produto(555) == 'bebidas'
assert verifica_tipo_do_produto(888) == 'brinquedos'
assert verifica_tipo_do_produto(0) is None
print('Função ok')
```

```python
df['nome_tipo_do_produto'] = df['tipo_do_produto'].apply(verifica_tipo_do_produto)
df
```

```python
for (nome, produto) in df[df['valido']].groupby([ 'nome_tipo_do_produto','nome_regiao_de_destino']):
    print (' %s que partem do %s '% (nome))
    display(produto)
```
